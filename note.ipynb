#### Thông tin state

dx là khoảng cách ngang của bird đến đến tâm lỗ

     Dương khi bird chưa đến tâm lỗ, âm khi đã qua tâm lỗ.
    - Tính khoảng cách đến cột tiếp theo chỉ khi đã qua hẳn. Nên có giá trị âm - khi vừa qua tâm lỗ mà chưa qua hẳn.
dy là khoảng cách dọc của bird đến tâm lỗ
    Dương khi bird trên tâm lỗ. Âm khi bird dưới tâm lỗ.
    Cũng cập nhật khoảng các đến tâm lỗ tiếp theo khi đã qua hẳn lỗ hiện tại

v là vận tốc của bird
    Dương khi bird rơi xuống. Âm khi bird đi lên.
    có giá trị từ -10 đến 10.
